```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use("fivethirtyeight")

full_data = pd.read_csv("C:\\Users\\barnw\\Documents\\A&M Classes\\UT Machine Learning\\Cardio_Project_1\\CardioGoodFitness.csv")
print(full_data)
product_data = full_data['Product'].dropna()
age_data = full_data['Age'].dropna()
gender_data = full_data['Gender'].dropna()
education_data = full_data['Education'].dropna()
marital_data = full_data['MaritalStatus'].dropna()
usage_data = full_data['Usage'].dropna()
fitness_data = full_data['Fitness'].dropna()
income_data = full_data['Income'].dropna()
miles_data = full_data['Miles'].dropna()
```

```python

sns.countplot(product_data)
plt.title("Product")
plt.xlabel("Product Type")
plt.ylabel("Quantity")
plt.show()
#It seems like the 195 sells the most, with numbers tapering off as we reach other (more advanced?) models.

age = plt.hist(age_data, bins = range(min(age_data), max(age_data) + 5, 5))
plt.title("Age")
plt.xlabel("Age (years)")
plt.ylabel("Quantity")
plt.show()
#The people who bought this product the most are relatively young, which makes sense because it is a fitness tool. It seems that most people who buy this are in the 20-30 year range.

sns.countplot(gender_data)
plt.title("Gender")
plt.xlabel("Gender Type")
plt.ylabel("Quantity")
plt.show()
#Men buy this more than women, but the difference does not seem extremely significant. We may want to market this slightly more towards men though.

sns.countplot(education_data)
plt.title("Education")
plt.xlabel("Education Length (years)")
plt.ylabel("Quantity")
plt.show()
#It seems that people with Associate's and Bachelor's degrees are most likely to buy this, and only those who managed to complete their degree as well. This implies that the moderately educated have a need for this product. However, this number drops when we get to the higher ends of the data. This could, however, be due to the scarcity of highly educated people in this study, since they are harder to come by in general. I would say that it is a safe bet to say that most educated people would buy this product.

sns.countplot(marital_data)
plt.title("Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Quantity")
plt.show()
#Although it is unclear whether there is a significant difference between partnered and single people, it is clear that the partnered people are more likely to buy this product based on whether this sample size alone.

sns.countplot(usage_data)
plt.title("Usage")
plt.xlabel("Usage (per week)")
plt.ylabel("Quantity")
plt.show()
#Most people seem to use this product three times a week, with four times being the second most used amount. This suggests that the people who use this exercise about half the time.

sns.countplot(fitness_data)
plt.xticks(np.arange(1, max(fitness_data)+1, 1))
plt.title("Fitness")
plt.xlabel("Fitness (Rating)")
plt.ylabel("Quantity")
plt.show()
#The people who use this seem to be moderately fit and active. With a fitness rating at its highest around 3, they are only somewhat fit, but not athletes.

income = plt.hist(income_data, bins = range(20000, max(income_data) + 1000, 1000))
plt.xticks(np.arange(20000, max(income_data)+15000, 15000))
plt.rc('xtick', labelsize=10)   
plt.title("Income")
plt.xlabel("Income (dollars)")
plt.ylabel("Quantity")
plt.show()
#As can be seen from this graph, the majority of people who buy this product have a somewhat lower end income, at around $50,000 annual income. This may mean that in general, this product should be advertised in lower income neighborhoods.

miles = plt.hist(miles_data, bins = range(0, max(miles_data) + 5, 5))
plt.title("Miles")
plt.xlabel("Miles")
plt.ylabel("Quantity")
plt.show()
#Most people run around 100 miles a week.
```

```python
# MULTIVARIATE ANALYSIS
sns.set(rc={'figure.figsize':(6.4,4.8)})
sns.stripplot(data=full_data, x="Income", y="Product", jitter=True)
plt.show()
#As can be seen by this data, the 195 and 498 are generally more lower end than the 798 model, so the 798 should only be marketed towards richer customers. Suggested to put ads and billboards in places that higher income customers frequent, such as upper-class neighborhoods.
sns.swarmplot(data=full_data, x="MaritalStatus", y="Product", size=3)
plt.show()
#It seems like single people are just as likely to buy each of these models as partnered people. However, it is interesting to note that the 195 is the most bought, perhaps due to its price or availability, especially since we noted that the 798 was mostly purchased by the wealthy.
sns.swarmplot(data=full_data, x="MaritalStatus", y="Gender", size=2.2)
plt.show()
#From this graph, we can see that partnered men are most likely to use this product, with single women the least likely.
sns.swarmplot(data=full_data, x="Gender", y="Product", size=2.2)
plt.show()
#We can now see that the 798 should not be marketed to women as it seems that they are not buying this model. However, numbers do not change at all in terms of men buying this product. Therefore, the cheaper model should be marketed to both genders, whereas the more expensive models should only be marketed to men.
sns.swarmplot(data=full_data, x="Usage", y="Gender", size=2.2)
plt.show()
#Men use this product a bit more than women, with women mostly using it less than four times a week and men using it four or more times.
sns.swarmplot(data=full_data, x="Usage", y="MaritalStatus", size=2.2)
plt.show()
#It seems that marital status has no effect on usage.
sns.stripplot(data=full_data, x="Age", y="Gender")
plt.show()
#This confirms what we already know about the fact that young men and women are the ones who use this product the most, so this product should be mostly marketed to them.
sns.set(rc={'figure.figsize':(9,4.8)})
sns.regplot(data=full_data, x="Miles", y="Usage")
plt.xticks(rotation = 45, ha = 'center')
plt.show()
sns.set(rc={'figure.figsize':(10,4.8)})
#The more people use this, the more miles they bike. No surprise there.
sns.regplot(data=full_data, x="Miles", y="Income")
plt.xticks(rotation = 45, ha = 'center')
plt.show()
#As income increases, so does the average number of miles that each person runs. This may be because of how much time people with higher income have as compared to lower income people.
sns.scatterplot(data=full_data, x="Age", y="Product")
plt.xticks(rotation = 45, ha = 'center')
plt.show()
#Young people and people over 50 seem to not use treadmills.
```
